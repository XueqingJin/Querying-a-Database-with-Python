# Part A – Query tickers with fetchall()

```python
#query tickers' id, ticker's name and ticker's company name, then returns a list of tuples

import sqlite3 as sql 
import pandas as pd
connection = sql.Connection("dj.db")
cursor = connection.cursor()
cursor.execute("SELECT id, ticker, name FROM tickers")
rows = cursor.fetchall()
df = pd.DataFrame(rows) 
df.columns = ['id', 'ticker', 'name']
df.set_index('ticker', inplace=True)
df
```

# Part B – Determine the date span of the data

```python
#query minimum date and maximum date in the table of daily_measures  

import sqlite3 as sql 
import pandas as pd
connection = sql.Connection("dj.db")
cursor = connection.cursor()
cursor.execute("SELECT min(date) as first, max(date) as last FROM daily_measures")
rows = cursor.fetchall()
print(rows)
```

# Part C – Query price history with for and Cursor

```python
#determine the date and closing price of the highest price for Apple stock during the dataavailable in the database.

import sqlite3 as sql
connection = sql.Connection("dj.db")
cursor = connection.cursor()
query = """
SELECT date, max(closing)
FROM daily_measures
WHERE ticker_id=3 
ORDER BY date
"""
cursor.execute(query)
rows = cursor.fetchall()
print(rows)
```

# Part D – Query price history to determine largest 1-day price percentage moves

```python
#determine the date and price when Apple had the highest one-day percent change. 

import sqlite3 as sql
connection = sql.Connection("dj.db")
cursor = connection.cursor()
query = """
SELECT date, closing, opening, max(high-low) as change
FROM daily_measures,tickers
WHERE name LIKE 'Apple%'
ORDER BY date
"""
cursor.execute(query)
rows = cursor.fetchall()
df = pd.DataFrame(rows) 
df.columns = ['date', 'closing','opening','change']
df
```
